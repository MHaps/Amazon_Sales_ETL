```python
#Data Insertion
import pandas as pd
import numpy as np
import psycopg2
df = pd.read_csv('z_amount_extracted.csv')

conn = psycopg2.connect(
    host="localhost",
    database="amazon_sales_report",
    user="postgres",
    password="password"
)
cur = conn.cursor()

for index, row in df.iterrows():
    amount = None if row['amount'] == "NULL" else float(row['amount'])
    try:
        cur.execute("INSERT INTO fact_sales (amount) VALUES (%s)",
        (amount,)
    )
    except Exception as e:
        print("ERROR:", e)
        conn.rollback()
    else:
        conn.commit()

cur.close()
conn.close()

print("Data inserted successfully.")
```

```python
import pandas as pd
import psycopg2
import numpy as np

df = pd.read_csv('z_amount_extracted.csv')

conn = psycopg2.connect(
    host="localhost",
    database="amazon_sales_report",
    user="postgres",
    password="password"
)
cur = conn.cursor()

for index, row in df.iterrows():
    amount_value = row['amount']
    if amount_value == 'NULL':
        amount_value = None
    cur.execute(
        "INSERT INTO salesmetrics (amount) VALUES (%s)",
        (amount_value,)
    )

conn.commit()

cur.close()
conn.close()

print("Data inserted successfully.")
```
